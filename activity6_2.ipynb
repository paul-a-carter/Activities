2. Create a dictionary with the word frequency for each word in the web page.

```python
# Your Code Here

fhand = urllib.request.urlopen('http://data.pr4e.org/romeo.txt')

counts = dict()

for line in fhand:
    words = line.decode().split()
    for word in words:
        counts[word] = counts.get(word,0) + 1

print(counts)

```
