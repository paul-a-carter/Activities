```python
# import packages 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#prepare data 
labels = ["a","b","c"]
values = [1,4,2]

# change bar colors 
colors = ["blue","red","yellow"]

#create bar chart
bars = plt.bar(labels,values,color=colors)

#(optional tweak) --> place hatch patterns on bars 
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

#save figure
plt.savefig("bar_chart.png",dpi=300)

#render plot 
plt.show()
```
